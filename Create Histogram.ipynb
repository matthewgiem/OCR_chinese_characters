```python
import glob
from matplotlib import pyplot as plt
import numpy as np
```

```python
dictionary_total = dict()
character_dictionary = dict()
for_histagram = []
for folder in glob.glob('tcc_bmp/*'):
    if folder[-4:] != '.bmp':
        place_holder = dict()
        place_holder_for_histagram = dict()
        jinwen = 0
        chuxi = 0
        oracle = 0
        smallseal = 0
        for file in glob.glob('{}/*'.format(folder)):
            if file[-4:] == '.bmp':
                if file.split('-')[1].split('_')[0] in dictionary_total:
                    dictionary_total[file.split('-')[1].split('_')[0]] += 1
                else:
                    dictionary_total[file.split('-')[1].split('_')[0]] = 1
                if file.split('-')[1].split('_')[0] in place_holder:
                    place_holder[file.split('-')[1].split('_')[0]] += 1
                else:
                    place_holder[file.split('-')[1].split('_')[0]] = 1
                if file.split('-')[1].split('_')[0] == 'jinwen':
                    jinwen +=1
                if file.split('-')[1].split('_')[0] == 'chuxi':
                    chuxi +=1
                if file.split('-')[1].split('_')[0] == 'oracle':
                    oracle +=1
                if file.split('-')[1].split('_')[0] == 'smallseal':
                    smallseal +=1
        for_histagram.append([['jinwen', jinwen],['chuxi', chuxi],['oracle', oracle],['smallseal', smallseal]])        
        character_dictionary[folder] = place_holder
                    
                    
                
                
```

```python
list_of_jinwen = []
for x in for_histagram:
    list_of_jinwen.append(x[0][1])
list_of_chuxi = []
for x in for_histagram:
    list_of_chuxi.append(x[1][1])
list_of_oracle = []
for x in for_histagram:
    list_of_oracle.append(x[2][1])
list_of_smallseal = []
for x in for_histagram:
    list_of_smallseal.append(x[3][1])
```

```python
bins = np.arange(0, 160, 10)
```

```python
plt.hist(list_of_jinwen, bins=bins, alpha=0.5, label='jinwen')
plt.hist(list_of_chuxi, bins=bins, alpha=0.5, label='chuxi')
plt.hist(list_of_oracle, bins=bins, alpha=0.5, label='oracle')
plt.hist(list_of_smallseal, bins=bins, alpha=0.5, label='smallseal')
plt.legend(loc='upper right')
plt.yscale('log', nonposy='clip')
plt.show()
```
